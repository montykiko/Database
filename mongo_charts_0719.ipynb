```python
city_list = []
for i in item_info.find({["area"][0]:"杭州"}):
    city_list.append(i['area'][1])
city_index = list(set(city_list))
print(city_index)
```

```python
post_times = []
for index in city_index:
    post_times.append(city_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    length = 0
    if length <= len(city_index):
        for area,times in zip(city_index,post_times):
            data = {
                'name':area,
                'data':[times],
                'type':types
            }
            yield data
            length += 1
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series, show='inline', options=dict(title=dict(text='杭州各地区二手物品发帖量')))
```
